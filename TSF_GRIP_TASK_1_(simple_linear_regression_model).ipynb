```python
#reading and printing the data
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('http://bit.ly/w-data')
data
```

```python
#plotting our data

data.plot(x="Hours",y="Scores",style=".")
plt.title('Hours VS Scores')
plt.ylabel('Scores')
plt.show()
```

```python
#prepare the data
x=data.iloc[:, :-1].values
y=data.iloc[:,1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
print("X_train:", X_train)
print("X_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)
```

```python
#train the model
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
#plot the regression line with the actual data
line = reg.coef_*x+reg.intercept_
plt.plot(x,line)
plt.scatter(x,y)
plt.title('Hours VS Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#test the model

predict=reg.predict(X_test)
df=pd.DataFrame({'Actual':y_test,'Preadicted':predict})
df
```

```python
#predicting with new data
hours=9.5
score = reg.predict([[hours]])
print("we predict score of {} with hours {} studied ".format(score[0],hours))
```

```python
#evaluate the model using MAE

from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,predict)
print("Mean absolute error = ", MAE)
```
